```python
import re

def price_processor(row):
  thousand = 1000
  lakh = 100000
  if 'Thousand' in row:
    return float(re.sub(r'[\s]|Thousand|thousand','', row)) * thousand

  return float(re.sub(r'[\s]|Lakh|thousand','', row)) * lakh



def string_processor(input_string):

  return int(re.sub(r'[,\s]|sqft', "", input_string)
)
```

```python
df['price'] = df['price'].apply(price_processor)
df['area'] = df['area'].apply(string_processor)

df['beds'] = df['beds'].str.replace(' Bed','').astype('int')
df['bath'] = df['bath'].str.replace(' Bath','').astype('int')

df['city'] = df['adress'].apply(lambda x: 'Dhaka' if 'Dhaka' in x else 'Chattogram')
```
